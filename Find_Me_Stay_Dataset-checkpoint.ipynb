```python
# weighted mean

def weighted_median(df, median_col , weight_col):
    df_sorted = df.sort_values(median_col)
    cumsum = df_sorted[weight_col].cumsum()
    cutoff = df_sorted[weight_col].sum()/2
    return df_sorted[cumsum >= cutoff][median_col].iloc[0]
weighted_median( data , "price_per_night" , "rating" )
```

```python
# variance - (how the data is spread) and estimate of location were to find the center of the data


from statistics import variance
variance(data["price_per_night"])
```

```python
# standard deviation - square root of variance

from statistics import stdev
stdev(data["price_per_night"])
```

```python
# mean absolute deviation - Mean of the absolute value of the deviation from the mean also known  as l1.norm or Manhattan Norm
# mean(absolute(X- mean(X))


from numpy import mean, absolute

mean(absolute(data["price_per_night"] - mean(data["price_per_night"])))

#mean absolute deviation ,mean and standard deviation ...all three are not robust to outliers
```

```python
# median absolute deviation (# median absolute deviation is robust to deviation)
# median(absolute(X- mean(X))


from numpy import median, absolute

median(absolute(data["price_per_night"] - median(data["price_per_night"])))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv("C:/Users/Admin/hotel_airnbn.csv")

# Group by 'location' and sum 'price_per_night'
category_data = data.groupby('location')['price_per_night'].sum()

# Plot the pie chart using all parameters
ax = category_data.plot.pie(
    figsize=(6, 6),                          # Size of the pie chart (width, height)
    autopct='%1.1f%%',                       # Format for percentage labels (1 decimal place)
    startangle=90,                           # Start angle for the first slice
    labels=category_data.index,              # Custom labels for the pie slices (location names)
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], # Custom colors for the slices
    pctdistance=0.75,                        # Percentage label distance from the center
    shadow=True,                             # Add shadow to the pie chart
    radius=1.2,                              # Increase radius of the pie chart
                                 
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5},  # Customize the edge of the slices
    textprops={'fontsize': 12, 'color': 'black'},  # Customize the text (label) style
    legend=None,                             # Display a legend
    use_index=True                           # Use index values as labels
)

# Add a title
ax.set_title('Price Per Night by Location')

# Remove the label (y-axis label) since it's not needed for pie charts
ax.set_ylabel('')

# Show the plot
plt.show()
```
